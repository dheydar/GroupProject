# LA Redlining Map

Within this section, we create a map of redlining within Los Angeles and the surrounding vicinity. This map is based on grades given to certain areas by the Home Owners' Loan Corporation in order to evaluate the lending risk within these areas. The redlining map outline is from the University of Richmond and within this section we overlap this map outline with the boundaries of LA County, a dataset from the LA Times.

```python
df = gpd.read_file('LA Redlining/cartodb-query.shp')
crs = {'init': 'epsg:4326'}
df.head()
#reads the redlining shape file and adjusts the coordinate reference system
```

```python
df.info()
```

```python
df.geometry
```

```python
df.plot(
            figsize=(20,12),   
            column = 'holc_grade',   
            legend = True,                
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1.3,1)
            }                  
) 
#allows you to adjust the redlining plot to include the HOLC grades in different shades, include a legend, and adjust the size of the plot
```

```python
neighborhoods = gpd.read_file('http://s3-us-west-2.amazonaws.com/boundaries.latimes.com/archive/1.0/boundary-set/la-county-neighborhoods-v5.geojson')
#loads the LA County boundaries from the LA Times
```

```python
neighborhoods = neighborhoods[['name','geometry']]
neighborhoods.head()
#reduces the columns within the LA County boundaries dataset 
```

```python
ax=neighborhoods.plot(figsize=(12,12),color='gainsboro', edgecolor='white')
#provides a plot of the LA County boundaries
```

```python
base = neighborhoods.plot(figsize=(12,10),color='gainsboro', edgecolor='white')
ax = df.plot(ax=base, column = 'holc_grade', markersize=5)
#Sets the base map as the LA County boundaries map and the top layer as the redlining map
```

```python
df.geometry.total_bounds
#shows you the bounding box coordinates for the redlining data 
```

```python
minx, miny, maxx, maxy = df.geometry.total_bounds
print(minx)
print(maxx)
print(miny)
print(maxy)
#puts the bounding box coordinates into their own variables
```

```python
base = neighborhoods.plot(figsize=(12,12),color='gainsboro', edgecolor='white')
ax = df.plot(ax=base, marker='o', column = 'holc_grade', markersize=5,  legend = True,              
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1.3,1)
            } )
ax.set_xlim(minx - .1, maxx + .1) 
ax.set_ylim(miny - .1, maxy + .1)
ax
#uses the bounding box coordinates to set the x and y limits, adds a legend, and zooms to the layer of the redlining data 
```
