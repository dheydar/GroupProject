# LA Redlining Map and Function

Within this section, we create a map of redlining within Los Angeles and the surrounding vicinity. This map is based on grades given to certain areas by the Home Owners' Loan Corporation in order to evaluate the lending risk within these areas. The redlining map outline is from the University of Richmond and within this section we overlap this map outline with the boundaries of LA County, a dataset from the LA Times. 

Followed by creating the redlining map of Los Angeles overall and zooming into the neighborhood level, a function is created to allow for the visualization of redlining within other neighborhoods in Los Angeles as well.

```python
# subset to a neighborhood
neighborhood = neighborhoods[neighborhoods.name=='Westwood']

# get the bounds
minx, miny, maxx, maxy = neighborhood.geometry.total_bounds

base = neighborhood.plot(figsize=(12,12),color='None', edgecolor='black')
ax = df.plot(ax=base, 
             column = 'holc_grade', 
             legend = True,
             alpha=0.5,
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1.3,1)
            } )
ax.set_xlim(minx - .01, maxx + .01) 
ax.set_ylim(miny - .01, maxy + .01)
ax
ax.set_title('LA Home Owners Loan Corporation Redlining Map, 1939', fontsize = 22)
#Adds a title to the previously created plot

#This code zooms in to a single neighborhood for redlining within Los Angeles
```

```python
def map_redlining_by_neighborhood(place='Westwood'):
    
    neighborhood = neighborhoods[neighborhoods.name==place]
    
    minx, miny, maxx, maxy = neighborhood.geometry.total_bounds
        
    base = neighborhood.plot(figsize=(12,12),color='None', edgecolor='black')
    
    ax = df.plot(ax=base, 
             column = 'holc_grade', 
             legend = True,
             alpha=0.5,
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1.3,1)
            } )

    ax.set_xlim(minx - .01, maxx + .01) 

    ax.set_ylim(miny - .01, maxy + .01)
                
    ax.axis('off')
                    
    ax.set_title('Redlining in '+neighborhood.name.values[0]+' Los Angeles',fontsize=20)
                
    ctx.add_basemap(ax,source=ctx.providers.CartoDB.Positron)
    ax

#this code is used to create a function for the redlining map in the Los Angeles area and zooms into the desired neighborhood
```
